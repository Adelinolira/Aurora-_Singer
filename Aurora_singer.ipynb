# Sistema de verificação de Decolagem

esse projeto simula um sistema de análise de telemetria usado antes da decolagem de uma nave.
o objetivo é analisar dados de sensores e verificar se todos os parâmetros estarão dentro das faixas seguras para permitir a decolagem.

caso algum valor seteja fora do limite seguro, o sistema aborta automaticamente a missão.

## 1.1 organização da telemetria
|parâmetro | Descrição | Faixa segura |
|-----------|-----------|-------------|
| Temperatura interna | Temperatura dentro da nave | 18 - 27 °C |
| Temperatura externa | Temperatura ambiente | -50 - 50 °C |
| Integridade estrutural | percentual da estrutura | >= 90% |
| Energia | Nível de carga | >= 70% |
| Pressão dos tanques | pressão do combustível | 80 - 120 bar |
| Status módulos críticos | funcionamento dos sistemas | OK |

Os dados de telemetria são coletados pelos sensores da nave e analisado antes da decolagem.
O sistema verifica se todos os valores estão dentro das faixas seguras para evitar riscos durante o lançamento.

## 1.2 Algoritmo de verificação

INÍCIO

ler a temperatura interna
ler a temperatura externa
ler integridade estrutural
ler nível de energia
ler a pressão dos tanques
ler status dos módulos

SE tempreratura interna fora da faixa segura       
   
   abortar

SE integridade estrutura < 90%
   
   abortar

SE energia < 70%
   
   abortar

SE pressão fora da faixa
   
   abortar

SE módulos != OK
   
   abortar

CASO CONTRÁRIO
   pronto para decolagem

fim

```python
# 1.3 Script

# Simulação de sensores

import random
import time

print("INICIANDO SISTEMA DE TELEMETRIA...")
time.sleep(2)
# ---------------------------
# Simulação dos sensores
# ---------------------------

temp_interna = random.randint(15, 30)
temp_externa = random.randint(-60, 60)
integridade = random.randint(85, 100)
energia = random.randint(60, 100)
pressao = random.randint(70, 130)
modulos = random.choice(["OK", "ERRO"])

print("\nColetando dados dos sensores...\n")
time.sleep(2)

print("Temperatuda interna:", temp_interna, "°C")
print("Temperatura externa:", temp_externa, "°C")
print("Integridade", integridade, "°C")
print("Energia disponível:", energia, "%")
print("Pressão dos tanques:", pressao, "bar")
print("Status dos módulos:", modulos)

time.sleep(3)

# ---------------------------
# Início das verificações
# ---------------------------

print("\nIniciando verificação dos sistemas...\n")
time.sleep(2)

pronto_para_decolagem = True

# Verificação da temperatura
print("Verificando temperatura interna...")
time.sleep(2)

if temp_interna < 18 or temp_interna > 27:
    print("Temperatuda fora do limite seguro")
    pronto_para_decolagem = False
else:
    print("temperatuda dentro do limite")
# Verificação da temperatura externa
print("\nVerificando temperatura externa...")
time.sleep(2)

if temp_externa < -50 or temp_externa > 50:
    print("Temperatuda fora da faixa de segurança")
    pronto_para_decolagem = False
else:
    print("Temperatuda dentro do limite de segurança")
# Verificação da energia
print("\nVerificando nível de energia...")
time.sleep(2)

if energia < 70:
    print("Energia insuficiente para decolagem")
    pronto_para_decolagem = False
else:
    print("Energia suficiente")
# Verificação de integridade
print("\nVerificando integridade estrutural...\n")
time.sleep(2)
if integridade < 90:
    print("Integridade estrutural abaixo do mínimo")
    pronto_para_decolagem = False
else:
    print("Integridade estrutural OK")
# verificação de pressão
print("\nVerificando pressão dos tanques...")
time.sleep(2)
if pressao < 80 or pressao > 120:
    print("Pressão fora da faixa segura")
    pronto_para_decolagem = False
else:
    print("Pressão dentro da faixa")
# verificação dos módulos
print("\nVerificando status dos módulos críticos...")
time.sleep(2)
if modulos != "OK":
    print("Falha nos módulos crítico")
    pronto_para_decolagem = False
else:
    print("Módulos funcionando corretamente")
time.sleep(2)
# ---------------------------
# Resuldato final
# ---------------------------
print("\nVerificações concluídas!\n")
time.sleep(3)

if pronto_para_decolagem:
    print("PRONTO PARA DECOLAR!")
else:
    print("DECOLAGEM ABORTADA")
```

```python
# 1.4 Verificação de Risco

print("\nAnalisando nível de risco da missão...")
if integridade < 85 or energia < 60:
  print("RISCO ALTO")
elif integridade < 92 or energia < 75:
  print("RISCO MODERADO")
else:
  print("RISCO BAIXO")
```

```python
# 1.5 análise energética

cap_max = 1000 # Capacidade máxima em kWh
carga_atual = 80 # Percentagem da carga atual (ex: 80%)
consumo = 500 # Consumo em kWh
perdas = 5 # Percentagem de perdas (ex: 5%)

#energia disponivel

energ_dispo = cap_max * (carga_atual / 100) # Faz o calcúlo da energia disponível em kWh

# energia após perdas

energ_apos_perdas = energ_dispo * (1 - (perdas / 100)) # Calcula energia após perdas em kWh

# energia restante após decolagem

energ_restante = energ_apos_perdas - consumo # Calcula energia restante em kWh

print(f"Energia disponível: {energ_dispo:.2f} kWh")
print(f"Energia após perdas: {energ_apos_perdas:.2f} kWh")
print(f"Energia restante após decolagem: {energ_restante:.2f} kWh")
```

# Ética

a responsabilidade das agências espaciais vai além do aspecto técnico. Ela envolver um compromisso com a preservação da vida humana e com futuro da exploração espacial. Esse compromisso pode ser entendido a partir de quatro pilares:

. Dever de cuidado (Preservação da vida): A segurança deve ser tratada como prioridade máxima. As agências têm a responsabilidade de evitar expor astronautas ou populações civis a riscos desnecessários em nome do progresso tecnológico ou econômico.

. Gestão de Bens Comuns (Sustentabilidade): O espaço é considerado um patrimônio da humanidade. Por isso, é importante que as missões sejam planejadas de forma a reduzir a geração de lixo espacial, preservando o ambiente orbital para futuras gerações.

.Transparência e Responsabilidade: as organizações envolvidas devem comunicar de forma clara os riscos das missões e assumir responsabilidade por eventuais danos causados, seja no espaço ou na Terra.

.Justiça Global: a exploração espacial deve gerar benefícios para toda a humanidade, e não apenas para países mais ricos. Além disso, falhas de segurança podem afetar infraestrutura globais, como satélites de comunicação e monitoramento climático.

Em resumo, a ética na exploração espacial exige que as atividade sejam conduzidas de maneira segura, sustentável e responsável, tratando o espaço como um ambiente que requer cooperação internacional e respeito à vida.

## Impacto social

O impacto social de segurança nas missões espaciais está diretamente ligado à confiança da sociedade e à continuidade da exploração tecnológica. Qaundo as agências espaciais demonstram altos níveis de segurança, o impacto tende a ser positivo; porém, quando ocorrem falhas, as consequencias podem ser significativas.

.Confiança e apoio público: O sucesso das missões fortalece o interesse da sociedade e o apoio ao investimento público em ciência e tecnologia. Por outro lado, acidentes históricos, como o desastre do Challenger, geram forte impacto emocional e podem levar à interrupção de programas espaciais porl ongos periodos.

.Democratização de acesso ao espaço: Com padrões de segurança cada vez mais rigorosos, a exploração espacial deixa de ser exclusiva de militares e cientistas, abrindo espaço para a participação de civis, pesquisadores e até turismo espacial.

.Proteção da infraestruturas essenciais: A segurança nas operações espaciais também protege satélites fundamentais para a sociedade moderna, responsáveis por serviços como GPS, comunicação, previsão do tempo e monitoramento de desastres naturais.

.Inspiração e educação: missões espaciais bem-sucedidas e seguras têm forte impacto educacional, inspirando jovens a seguir carreiras nas áreas de ciência, tecnologia, engenharia e matemática, contribuindo para o desenvolvimento científico e econômico.

Em resumo, a segurança nas missões espaciais transforma a exploração do espaço em um fator de desenvolvimento científico, tecnológico e social para humanidade.

## Sustentabilidade

No contexto da exploração espacial, a sustentabilidade é um dos pilares fundamentais para garantir que o espaço continue sendo utilizado pelas próximas gerações. Nesse sentido, a segurança das operações espaciais também está diretamente relacionada à preservação do ambiente orbital.

.Preservação das órbitas: uma das maiores preocupações é evitar a chamada Síndrome de Kessler, uma reação em cadeia que colisões entre objetos geram grandes quantidades de detritos espaciais, podendo tornar determinadas órbitas inutilizáveis por longos períodos. Por isso, manter o espaço orbital livre de lixo é uma responsabilidade importante das agências espaciais.

.Gestão do cliclo de vida dos satélites: missões sustentáveis planejam o destino final dos equipamentos espaciais. Isso pode incluir a reentrada conrolada na atmosfera ou o envio para chamadas "órbitas cemitério", evitando que satélites desativados se tornem detritos perigosos.

.Proteção planetária: também existe a preocupação de não contaminar outros corpos celestes, como Marte ou a Lua, com microrganismos da Terra. Essa medida preserva a integridade científica desses ambientes para futuras pesquisas.

.Eficiência no uso de recursos: O desenvolvimento de tecnologia reutilizáveis, como foguetes caoazes de pousar e ser utilizado novamente, contribui para reduzir desperdícios e diminuir o impacto ambienta dos lançamentos.

Em síntese, a sustentabilidade espacial busca garantir que a exploração do espaço seja realizada de forma responsável, preservando o ambiente orbital e permitindo que gutuas gerações também possam utilizá-los.
